```python

#3...

import keras
#import keras backend
import keras.backend as K
# f1-score function
def aad_f1_multiclass_3_convert_3out_to1(y_true, y_pred): #taken from old keras source code

    y_true= tf.cast(y_true, tf.int32)

#----------------- 3 out from last dense layer to 1 ----> not for nearest neighbour, ibl( only keras)
    y_pred= tf.argmax(y_pred, axis=-1)

    rnd= tf.round(y_pred)
    rnd= tf.cast(rnd, tf.int32)

    sum_pred_true = tf.add(rnd,y_true)    #y_mul= rnd * y
    mult_pred_true = tf.multiply(rnd,y_true)

    const_0 = tf.constant(0)
    const_2 = tf.constant(2)
    const_4 = tf.constant(4)

    eq_0_bool = tf.math.equal(sum_pred_true, const_0)
    eq_2_bool= tf.math.equal(sum_pred_true, const_4)
    eq_0_int= tf.cast(eq_0_bool, tf.int32)
    eq_2_int= tf.cast(eq_2_bool, tf.int32)
    tp_0=tf.reduce_sum(eq_0_int)
    tp_2=tf.reduce_sum(eq_2_int)

    const_1 = tf.constant(1)
    eq_1_bool = tf.math.equal(mult_pred_true, const_1)
    eq_1_int= tf.cast(eq_1_bool, tf.int32)
    tp_1=tf.reduce_sum(eq_1_int)

    tot_wrong_pred = tf.reduce_sum(tf.cast( tf.not_equal (rnd, y_true ),  tf.int32  )  )

    f1_0 = (const_2* tp_0 ) / (  (const_2* tp_0 ) + tot_wrong_pred  )
    f1_1 = (const_2* tp_1 ) / (  (const_2* tp_1 ) + tot_wrong_pred  )
    f1_2 = (const_2* tp_2 ) / (  (const_2* tp_2 ) + tot_wrong_pred  )
#- - - -> 2D
    f1_avg = ( f1_0+ f1_1+ f1_2)/3

    return f1_avg
print("yeah")
```

```python
import keras

#import keras backend
import keras.backend as K

# f1-score function
def get_f1(y_true, y_pred): #taken from old keras source code
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision + recall + K.epsilon())
    return f1_val
```

```python
import keras

#import keras backend
import keras.backend as K

# f1-score function
def get_f1(y_true, y_pred): #taken from old keras source code
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision + recall + K.epsilon())
    return f1_val
```
